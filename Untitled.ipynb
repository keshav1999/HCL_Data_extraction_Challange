```python
def get_stattingtop(lis):
    n = len(lis)
    start =0
    for i in range(n):
        k = lis[i].upper()
        if k.find("BALANCE SHEET")!=-1:
            start = i
            break
    return start

def get_line_coordinates(s):
    count = 0
    start = 0
    end = 0
    lis = []
    st = get_stattingtop(s)
    for i in range(len(s)):
    
        if s[i]==" ":
            lis.append(0)
        else:
            lis.append(1)
    liss0 = []
    liss1 = []
    count1 = 0
    count0 = 0
    for i in lis:
        if i==0:
            count0 = count0+1
            if count1>0:
                liss1.append(count1)
                count1=0
        else:
            count1 = count1+1
            if count0>0:
                liss0.append(count0)
                count0=0
    liss0.append(count0)
    return liss0
        
def get_ending(data):
    n = len(data)
    sta = get_stattingtop(data)
    for i in range(sta+1,len(data)):
        lin = data[i]
        if lin.find("STATEMENTS")!=-1:
            return i
        cord = get_line_coordinates(lin)
        
        count=0
        for j in cord:
            if j==1:
                count+=1
        if count>8:
            return i
            
        
    
```
